# Apriori Implementation

```python
df.head(1)
```

```python
data = df.copy()
data1 = data.copy()
```

```python
newdata = pd.get_dummies(data['itemDescription'])
```

```python
newdata.head(2)
```

```python
data1.drop('itemDescription', axis=1, inplace=True)
```

```python
data1.head()
```

```python
data1 = data1.join(newdata)
```

```python
data1.head()
```

```python
products = df.itemDescription.unique()
```

```python
data1.groupby(['Member_number', 'Date'])[products[:]].sum()
```

```python
newitemsdf = data1.groupby(['Member_number', 'Date'])[products[:]].sum()
```

```python
newitemsdf.head(2)
```

```python
newdata
```

```python
newitemsdf = newitemsdf.reset_index()[products]
```

```python
newitemsdf
```

```python
def func(data):
  for product in products:
    if data[product] > 0:
      data[product] = product
  return data
```

```python
newitemsdf = newitemsdf.apply(func, axis=1)
```

```python
newitemsdf
```

```python
newitemsdf.values[0]
```

```python
['whole milk','yogurt', 'sausage','semi-finished bread']
```

```python
arr = []

for i in range(newitemsdf.shape[0]):
  item_name = []
  for j in newitemsdf.loc[i]:
    if j != 0:
      item_name.append(j)
  arr.append(item_name)
  item_name = []
```

```python
arr[1]
```

```python
arr
```

```python
association_model = apriori(arr)
result = list(association_model)
```

```python
result
```

```python
association_model = apriori(arr,
                            max_length=2,
                            min_confidence=0.10,
                            min_support=0.0001)
result = list(association_model)
```
